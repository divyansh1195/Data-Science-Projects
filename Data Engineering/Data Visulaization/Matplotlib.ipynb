## Matplotlib

**Scipy**:\
We have the min and max temperatures in a city In India for each months of the year.
We would like to find a function to describe this and show it graphically, the dataset
given below.

Task:
1. fitting it to the periodic function
2. plot the fit

Data\
Max = 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25\
Min = 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18

```python
import numpy as np
import matplotlib.pyplot as plt

Max = np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
Min = np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])

months = np.arange(12)
plt.plot(months, Max, 'ro')
plt.plot(months, Min, 'bo')
plt.xlabel('Month')
plt.ylabel('Max and Min Temperatures')
```

```python
from scipy import optimize
def yearly_temps(times, avg, ampl, time_offset):
    return (avg + ampl * np.cos((times + time_offset) * 2 * np.pi / times.max()))

res_max, cov_max = optimize.curve_fit(yearly_temps, months,
                                      Max, [20, 11, 0])
res_min, cov_min = optimize.curve_fit(yearly_temps, months,
                                      Min, [-30, 30, 0])
```

```python
days = np.linspace(0, 12)

plt.figure()
plt.plot(months, Max, 'ro')
plt.plot(days, yearly_temps(days, *res_max), 'r-')
plt.plot(months, Min, 'bo')
plt.plot(days, yearly_temps(days, *res_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```

```python
#Loading Data and modules
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
url="https://raw.githubusercontent.com/BigDataGal/Python-for-Data-Science/master/titanic-train.csv"
titanic = pd.read_csv(url)
titanic.columns
```

**Create a pie chart presenting the male/female proportion**

```python
prop=round((titanic['Sex'].value_counts())/len(titanic)*100,2)
pd.DataFrame(prop)#check proportion wnumerically
```

```python
#Pie-Chart
labels = ['Male','Female']
sex_prop = titanic.Sex.value_counts()
fig1, ax1 = plt.subplots()
ax1.pie(sex_prop, labels=labels, autopct='%1.1f%%', shadow=True, colors = ['Yellow'
,'Cyan'])
plt.show()
```

**Create a scatterplot with the Fare paid and the Age, differ the plot color by gender**

```python
plt.figure()
gender = titanic[titanic.Sex=='male'].plot.scatter('Age', 'Fare', 
                                                   color='cyan',label='male')
titanic[titanic.Sex=='female'].plot.scatter('Age', 'Fare',
                                            color='RED',label='female',ax=gender)
```
